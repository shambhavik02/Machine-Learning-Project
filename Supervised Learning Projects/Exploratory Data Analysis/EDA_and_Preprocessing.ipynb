# Exploratory Data Analysis & Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set_theme(style="whitegrid")
```

## 1. Data Loading

```python
# Load the built-in Titanic dataset from seaborn
df = sns.load_dataset('titanic')
display(df.head())
```

## 2. Basic Data Information

```python
print("Dataset Shape:", df.shape)
print("\nData Types and Missing Values:")
df.info()
```

```python
print("Statistical Summary of Numerical Columns:")
display(df.describe())
```

## 3. Data Cleaning & Preprocessing

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values[missing_values > 0])
```

```python
# Handle missing values purely analytically
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df.drop(columns=['deck'], inplace=True) # Too many missing values

print("\nMissing values after cleaning:")
print(df.isnull().sum().max()) # Should be 0
```

## 4. Data Visualization & Analysis

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='class', hue='survived', palette='Set2')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Compute correlation matrix for numerical features
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```
